```python
import requests
import pandas as pd
import json

#导入需要用到的库#
def housecounts(url):#抓取网站数据，唯一参数是网页的URL（原网页随着时间的变化而更新），URL可写死#
    res = requests.post(url)
    jd = json.loads(res.text)
    #抓到数据并解析成字典#    
    
    def getHousedetails(i):
        results= {}
        results['name']=jd['rows'][i]['PROJECTALL_NAME']#项目名称#
        results['count']=jd['rows'][i]['TS']#已售套数#
        return results
       #解析单个数据#
    
    number = len(jd['rows'])
    #统计成交项目总数#    
    
    houses=[]
    for p in range(number):
        houses.append(getHousedetails(p))
    #所有成交项目名称、成交套数 # 
    
    df = pd.DataFrame(houses)
    #转换成DataFrame格式#
    df.to_csv('house.csv')   
    #保存到csv文件#
    print(df)
    return df
```

```python
url = 'http://www.zstmsf.com/website/app/FAffairs/GetJrcjqk/'
housecounts(url)
```
